```python
df = pd.read_csv('countries.csv')
df
```

```python
# 1. China and USA data
```

```python
us = df[df.country == 'United States']
us
```

```python
ch = df[df.country == 'China']
ch
```

```python
#Comparison part
```

```python
plt.plot(us.year, us.population)
plt.show()
```

```python
plt.plot(ch.year, ch.population)
plt.show()
```

```python
plt.plot(us.year, us.population)
plt.plot(ch.year, ch.population)
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['US data', 'China data'])
plt.title('Analysis of Population')
plt.show()
```

```python
plt.plot(us.year, us.population,'d')
plt.plot(ch.year, ch.population,'*')
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['US data', 'China data'])
plt.title('Analysis of Population')
plt.show()
```

```python
plt.plot(us.year, us.population/10**6,'p')
plt.plot(ch.year, ch.population/10**6)
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['US data', 'China data'])
plt.title('Analysis of Population')
plt.show()
```
