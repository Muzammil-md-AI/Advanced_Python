```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# 5 Countries popuation growth 
```

```python
df = pd.read_csv('countries.csv')
df
```

```python
In = df[df.country == 'India']

In
```

```python
USA = df[df.country == 'United States']

USA
```

```python
Indo = df[df.country == 'Indonesia']

Indo
```

```python
Nz = df[df.country == 'New Zealand']

Nz
```

```python
Ch = df[df.country == 'China']

Ch
```

```python
# Comparison analysis
```

```python
plt.plot(In.year, In.population/10**6)
plt.plot(USA.year, USA.population/10**6)
plt.plot(Indo.year, Indo.population/10**6)
plt.plot(Nz.year, Nz.population/10**6)
plt.plot(Ch.year, Ch.population/10**6)
plt.title('Population analysis')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['India','USA','Indonesia','New Zealand','China'])
plt.show()
```

```python
plt.plot(In.year, In.population/10**6,'o')
plt.plot(USA.year, USA.population/10**6,'*')
plt.plot(Indo.year, Indo.population/10**6,'d')
plt.plot(Nz.year, Nz.population/10**6,'x')
plt.plot(Ch.year, Ch.population/10**6)
plt.title('Population analysis')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['India','USA','Indonesia','New Zealand','China'])
plt.show()
```
